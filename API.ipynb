```python
import requests
import json
from bs4 import BeautifulSoup
from tqdm import tqdm
import pandas as pd
from time import sleep
```

```python
#так как API работает со словарем, создаем список и словарь для хранения ключей книг и их данных
links = []
works_data = {}

#собираем 500 книг при помощи цикла (5 запросов по 100 книг <как я поняла, больше нельзя>)
for offset in range(0, 500, 100):
    url = f"https://openlibrary.org/subjects/history.json?details=false&limit=100&offset={offset}"
    article = requests.get(url)
    articles = json.loads(article.text)['works']
    for a in articles:
        links.append(a['key']) #сохраняем ключ
        works_data[a['key']] = a #сохраняем данные по ключу

print(len(links))
```

```python
#создаем список для хранения информации по каждой книге
info = []

for l in tqdm(links):
    a = works_data[l]  #берём уже сохранённые данные
    name = a['title']
    author = a['authors'][0]['name']
    cover_id = a.get('cover_id')
    picture = f"https://covers.openlibrary.org/b/id/{cover_id}-M.jpg"
    year = a.get('first_publish_year')
    number = a['edition_count']
    link = "https://openlibrary.org" + l #извлекли все необходимые данные и сформировали ссылку
    info.append([name, author, picture, year, number, link])
    sleep(1)

print(info)
```

```python
#создаем датафрейм со всей информацией
df = pd.DataFrame(info, columns=['Название', 'Автор', 'Обложка', 'Год первого издания', 'Кол-во изданий', 'Ссылка'])
df
```
